### C13. Concaténation
Concaténez horizontalement les moyennes par colonne avec un array de zéros.


```python
# Votre code ici
zeros_array=np.zeros((1,4))
column_means=np.mean(features,axis=0).reshape(1,4)
result=np.hstack((zeros_array,column_means))
print(result)
```
