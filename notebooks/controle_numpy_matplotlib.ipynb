# Partie D — Visualisation avec Matplotlib (Plots Simples)

### D17. Line plot
Line plot des longueurs des sépales pour les 50 premières fleurs.


```python
# Votre code ici
import matplotlib.pyplot as plt
sepal_length=features[:50,0]
plt.figure(figsize=(8,4))
plt.plot(sepal_length,marker='o',linestyle='-')
plt.title('longueurs des sépales pour les 50 premières fleur')
plt.xlabel("index de fleur")
plt.ylabel("sepal length")
plt.grid(True)
plt.show()
```

### D18. Scatter plot

```python
# Votre code ici
petal_length=features[:,2]
petal_width=features[:,3]
plt.figure(figsize=(6,6))
plt.scatter(petal_length,petal_width,c='blue',alpha=0.6)
plt.title('longueur vs largeur des pétales')
plt.xlabel("longueur des pétales")
plt.ylabel("largeur des petales ")
plt.grid(True)
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
# Votre code ici
setosa= features[:50,:]
versicolor=features[50:100,:]
virginica = features[100:150,:]

means=[np.mean(setosa[:,0]),np.mean(versicolor[:,0]),np.mean(virginica[:,0])]
stds=[np.std(setosa[:,0]),np.std(versicolor[:,0]),np.std(virginica[:,0])]

plt.figure(figsize=(6,5))
plt.bar(['setosa','versicolor','virginica'],means,yerr=stds,capsize=5,color=['green','blue','red'])
plt.ylabel("Sepla length")
plt.title("moyenne de la longueur des sepales par espece avec ecart-type")
plt.show()
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
# Votre code ici
setosa_width=features[:50,3]
versicolor_width=features[50:100,3]
virginica_width=features[100:150,3]

setosa_cumsum=np.cumsum(setosa_width)
versicolor_cumsum=np.cumsum(versicolor_width)
virginica_cumsum=np.cumsum(virginica_width)

plt.figure(figsize=(8,5))
plt.plot(setosa_cumsum,label='setosa',color='green')
plt.plot(versicolor_cumsum,label='vericolor',color='blue')
plt.plot(virginica_cumsum,label='virginica',color='red')

plt.title("Somme cumulee des largeur des petales par espece")
plt.xlabel("index de fleurs")
plt.ylabel("cumsum petal width")
plt.legend()
plt.grid(True)
plt.show()
```
